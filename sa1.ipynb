```python
import numpy as np
import matplotlib.pyplot as plt

# Simple simulation of Open Loop and Closed Loop Power Control

# Number of users
n = 10

# Distance of users from base station (in km)
d = np.linspace(0.5, 5, n)

# Path loss exponent
alpha = 3

# Transmitted power
Pt = 1  # in Watts

# -------------------------------
# Open Loop Power Control
# -------------------------------
Pr_open = Pt / (d**alpha)

# -------------------------------
# Closed Loop Power Control
# -------------------------------
target_power = 0.05  # Desired received power

Pr_closed = np.zeros(n)

for i in range(n):
    # Adjust transmitted power according to distance
    Pt_adjusted = target_power * (d[i]**alpha)

    # Received power after adjustment
    Pr_closed[i] = Pt_adjusted / (d[i]**alpha)

# -------------------------------
# Display Results
# -------------------------------
print('Distance (km)    Open Loop Power    Closed Loop Power')
for i in range(n):
    print(f'{d[i]:<15.2f} {Pr_open[i]:<19.4f} {Pr_closed[i]:<19.4f}')

# -------------------------------
# Plotting
# -------------------------------
plt.figure()

plt.plot(d, Pr_open, 'r-o', linewidth=2, label='Open Loop Power Control')
plt.plot(d, Pr_closed, 'b-*', linewidth=2, label='Closed Loop Power Control')

plt.grid(True)

plt.xlabel('Distance from Base Station (km)')
plt.ylabel('Received Power')

plt.title('Analysis of Power Control Techniques')

plt.legend()
plt.show()
```
