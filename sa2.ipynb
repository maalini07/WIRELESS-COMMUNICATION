```python
import matplotlib.pyplot as plt
import pandas as pd


Pt = 30
Gt = 15
Gr = 15
L = 120
Pr_min = -80
Pr = Pt + Gt + Gr - L

Fade_Margin = Pr - Pr_min

print(f'Received Power = {Pr:.2f} dBm')
print(f'Receiver Sensitivity = {Pr_min:.2f} dBm')
print(f'Fade Margin = {Fade_Margin:.2f} dB')

if Fade_Margin > 20:
    print('Excellent Link Reliability')
elif Fade_Margin > 10:
    print('Good Link Reliability')
else:
    print('Poor Link Reliability')

plot_data = pd.DataFrame({
    'Metric': ['Received Power', 'Fade Margin'],
    'Value': [Pr, Fade_Margin]
})

plt.figure(figsize=(8, 6))
plt.bar(plot_data['Metric'], plot_data['Value'], color=['skyblue', 'lightcoral'])

plt.ylabel('Value in dB')
plt.title('Fade Margin Estimation in Wireless Link')

plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
