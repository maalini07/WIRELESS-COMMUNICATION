```python
import numpy as np
import matplotlib.pyplot as plt


N = 1000

signal = np.random.randn(N)

noise1 = 0.5 * np.random.randn(N)
noise2 = 0.5 * np.random.randn(N)

r1 = signal + noise1
r2 = signal + noise2


signal_power = np.mean(signal**2)

noise_power1 = np.mean(noise1**2)

SNR_before = 10 * np.log10(signal_power / noise_power1)


r_combined = (r1 + r2) / 2

combined_noise = r_combined - signal

noise_power_combined = np.mean(combined_noise**2)

SNR_after = 10 * np.log10(signal_power / noise_power_combined)


print(f'SNR Before Diversity = {SNR_before:.2f} dB')
print(f'SNR After Diversity  = {SNR_after:.2f} dB')
print(f'SNR Improvement      = {SNR_after - SNR_before:.2f} dB')


plt.figure(figsize=(8, 6))

snr_values = [SNR_before, SNR_after]
labels = ['Before Diversity', 'After Diversity']
plt.bar(labels, snr_values, color=['skyblue', 'lightcoral'])

plt.ylabel('SNR (dB)')
plt.title('SNR Improvement Using Diversity Techniques')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
